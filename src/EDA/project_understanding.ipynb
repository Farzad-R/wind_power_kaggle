```python
import pandas as pd
from pyprojroot import here

import warnings
warnings.filterwarnings('ignore')
```

```python
def convert_date_to_timestamp(df):
    df["timestamp"] = df['date'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d%H'))
    df = df.set_index("timestamp")
    df = df.drop(columns=["date"])
    df = df.sort_index()
    return df
```

```python
df_train = pd.read_csv(here("data/raw/train.csv"))
df_train = convert_date_to_timestamp(df_train)
df_train = df_train.sort_index()

print("train:", df_train.shape)

df_test = pd.read_csv(here("data/raw/test.csv"))
df_test = convert_date_to_timestamp(df_test)
df_test = df_test.sort_index()
print("test:", df_test.shape)

df_wf1 = pd.read_csv(here("data/raw/windforecasts_wf1.csv"))
df_wf1 = convert_date_to_timestamp(df_wf1)
df_wf1 = df_wf1.sort_index()

print("wf1:",df_wf1.shape)
```

```python
print("train first valid day:", df_train.first_valid_index(), "|", "train last valid day:", df_train.last_valid_index())
print("\n")

print("wind speed first valid day:", df_wf1.first_valid_index(), "|", "wind speed last valid day:", df_wf1.last_valid_index())
print("\n")

print("test first valid day:", df_test.first_valid_index(), "|", "test last valid day:", df_test.last_valid_index())
```
